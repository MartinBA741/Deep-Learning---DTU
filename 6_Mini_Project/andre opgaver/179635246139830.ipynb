Furthermore, the `batch_generator` also handles validation splitting.

Choose a reasonable batch size the `batch_size = ?` parameter below, and note your choice.

 * **Answer:**
 If the batch size is to big, there will be problems with memory, as there will be problems with fetching the data. However, if the batches are too small, the estimates of the gradients will not be as correct and as a result, the training speed will be very slow.
